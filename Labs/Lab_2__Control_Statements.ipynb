```python
### indentation :
### control statement
## to check if a number is positive , negative or zero
num = int(input("Enter your number:"))
if num>0:
  print("The number is positive")
```

```python
### whether the number is POSITIVE or NEGATIVE

num = int(input("Enter your number:"))
if num>0:
  print("The number is positive")
elif num<0:
  print("The number is negative")
elif num==0:
  print("The number is zero")
```

```python
### whether the given number is ODD or EVEN

mynum = int(input("Enter the number:"))

if mynum %2==0:
  print("The number is even")
else:
  print("The number is odd")
```

```python
### calculate the total sum of numbers

Cal = [89,90,78,77,88,67,65,89]
sum = 0
for i in Cal:
  sum = sum+i
  print(" Total marks: ", sum)
```

```python
### condition
for a in Cal:
  if a>87:
    print(a)
```

```python
### python code to print multiplication table

for i in range(1,11):
  for j in range(1,11):
    print("{:5}".format(i*j),end="")
  print()
```

```python
### while Loop

count = 0
while(count<5):
  print("My count is ", count)
  count = count+1
```

```python
Students = ("Vasanth","Mahendra","Harsha","Jagadish","Keshav")

for i in Students:
  if len(i)>6:
    print(i)
```

```python
### Loops in the dictionary

student = {"Vasanth":100, "Jagadish":98, "Keshav":95, "Venky":94}
for i in student:
  marks = student[i]
  #print (marks)
  print ("Student %s has scored %d " %(i, marks))
```

```python
nums = [0,1,2,3,4,5]
sq_n =[]
for i in nums:
  sq_n.append(i**2)
print(sq_n)
```

```python
Main_list = [[1,2,3],[4,5,6],[7,8,9]]
```

```python
Main_list
```

```python
Consise_List =[]
for sublist in Main_list:
  # print( sublist)
  for el in sublist:
    Consise_List.append(el)
print(Consise_List)
```

```python
print([ele for slist in Main_list for ele in slist])  ### comprehension
```

```python
planets_nested = [['Mercury', 'Venus','Earth'],
                 ['Mars','Jupiter','Saturn'],
                 ['Uranus','Neptune','Pluto']]

# Filter planets with length less than 6
filtered_planets = [planet for sublist in planets_nested for planet in sublist if len(planet) < 6]

print(filtered_planets)

# Expected Output:
#     ['Venus','Earth','Mars','Pluto']
```
